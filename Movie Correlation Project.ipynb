```python
# Import necessary libraries 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create

pd.options.mode.chained_assignment = None

# read in the data 

df = pd.read_csv(r'/Users/mollygorman/Downloads/movies.csv')
```

```python
# looking at the data 
df.head()
```

```python
# checking if there is any missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing)*100))
```

```python
# Data types for columns 

df.dtypes
```

```python
# Change data type of columns to remove any excess decimals such as '.0' to clean the look of the data 


df['budget'] = df['budget'].fillna(0).astype('int64')

df['gross'] = df['gross'].fillna(0).astype('int64')

df['votes'] = df['votes'].fillna(0).astype('int64')

df['runtime'] = df['runtime'].fillna(0).astype('int64')

```

```python
# Drop any duplicates, and NA values, as well as extract the year to clean the data for analysis

 # no duplicates

 #df = df.dropna()

#df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)


```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending= False)
```

```python
# Hypothesis - There will be a high correlation between budget and gross revenue
# The more money they spend, then there will be more revenue returned 
#Company will also have high correlation, the bigger the company the more revenue will come back in 
```

```python
# Scatter plot, budget vs gross revenue 


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Revenue')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Revenue')

plt.show()
```

```python
# Plot budget vs gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'blue'}, line_kws={'color':'green'})
```

```python
# Company vs revenue
# Taking the string company and assigning it numerically so we can use in correlation matrix
```

```python
df_numbered = df

for col_name in df_numbered.columns:
    if(df_numbered[col_name].dtype == 'object'):
        df_numbered[col_name] = df_numbered[col_name].astype('category')
        df_numbered[col_name] = df_numbered[col_name].cat.codes
        
df_numbered
```

```python
correlation_matrix = df_numbered.corr(numeric_only =True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
corr_matrix = df_numbered.corr()

corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
